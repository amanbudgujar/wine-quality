```python
import statsmodels.api  as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = dfred[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
Y = dfred['quality']
```

```python
X[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 
      'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
      'pH', 'sulphates', 'alcohol']] = scale.fit_transform(X[['fixed acidity', 
    'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
    'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 
                                                              'alcohol']])
```

```python
print(X)
```

```python
est = sm.OLS(Y,X).fit()
est.summary()
```

After observing the p-values from the above summary we can conclude that *volatile acidity* , *sulphates* and *alcohol* have the highest correlation with the quality of the wine.

*Creating a linear regression model that takes volatile acidity , sulphates and alcohol as independant variables*

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
lr = LinearRegression()
```

```python
X = dfred[['volatile acidity','sulphates', 'alcohol']]
y = dfred['quality']
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y
```

```python
line = lr.coef_*X+lr.intercept_

# Plotting for the test data
plt.scatter(X,y)
plt.plot(X, line);
plt.show()
```
